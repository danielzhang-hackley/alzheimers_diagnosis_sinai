```python
import pandas as pd
```

```python
expression_data = pd.read_csv(r'../data/expression_data', header=None, low_memory=False)
phenotype_data = pd.read_csv(r'../data/phenotype_data')

expression_data = expression_data.transpose()
expression_data.columns = expression_data.iloc[0]
expression_data = expression_data[1:]
expression_data.reset_index(level=0, inplace=True)
expression_data.drop('index', inplace=True, axis=1)

phenotype_data.drop('patient', inplace=True, axis=1)
```

```python
patients = pd.merge(expression_data['seqLibID'], phenotype_data['seqLibID'], how='outer')
sex = phenotype_data.drop(['DiagnosisByCognition', 'AgeOfDeath'], axis=1, inplace=False)
aod = phenotype_data.drop(['DiagnosisByCognition', 'sex'], axis=1, inplace=False)
cog = phenotype_data.drop(['AgeOfDeath', 'sex'], axis=1, inplace=False)

X = pd.merge(patients, expression_data, on='seqLibID', how='outer')
X = pd.merge(X, sex, on='seqLibID', how='outer')

y_aod = pd.merge(patients, aod, on='seqLibID', how='outer')
y_cog = pd.merge(patients, cog, on='seqLibID', how='outer')

X.to_csv(r'../data/X.csv')
y_aod.to_csv(r'../data/y_aod.csv')
y_cog.to_csv(r'../data/y_cog.csv')
```
